<h1 style="color:#7d19ff">Contexto</h1>

Uma aplicação que facilita o acesso aos dados de Mh2 de forma offline, dado que o game não é muito popular as fontes de dados desse game geralmente são encontradas incompletas ou com atualizações muito velhas. Nessa aplicação deve conter o acesso de maneira offline aos seguintes dados:

    - 1.1. All Quests.
    - 1.2. All Monsters.
    - 1.3. All Itens.
    - 1.4. All Kitchen Combinations.
    - 1.5. Mix Sets.

<h1 style="color:#7d19ff">Desafio</h1>

<h2>Problema</h2>
<pre>◈ Falta de dados com fácil acesso de maneira offline.</pre>

<h2>Causa</h2>
<pre>◈ Dificuldade de acessar determinados dados.
    ◈ Dados de outras fontes incompletos e sem atualizações.</pre>
