<h1 style="color:#7d19ff">Contexto</h1>

Uma aplicao que facilita o acesso aos dados de Mh2 de forma offline, dado que o game no  muito popular as fontes de dados desse game geralmente so encontradas incompletas ou com atualizaes muito velhas. Nessa aplicao deve conter o acesso de maneira offline aos seguintes dados:

    - 1.1. All Quests.
    - 1.2. All Monsters.
    - 1.3. All Itens.
    - 1.4. All Kitchen Combinations.
    - 1.5. Mix Sets.

<h1 style="color:#7d19ff">Desafio</h1>

<h2>Causa</h2>

     Dados de outras fontes incompletos e sem atualizaes.</pre>
