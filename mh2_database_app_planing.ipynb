<h1 style="color:#7d19ff">Contexto</h1>

<pre> Uma aplicao que facilita o acesso aos dados de Mh2 de forma offline, dado que o game no  muito popular as fontes de dados desse jogo geralmente so encontradas geralmente incompletas ou com atualizaes muito velhas.
Nessa aplicao deve conter um aesso fcil aos seguintes dados de forma offline:</pre>

<pre>    1.1. All Quests.
    1.2. All Monster.
    1.3. All Items.
    1.4. All Kitchen Combinations.
    1.5. Mix Sets.</pre>

<h1 style="color:#7d19ff">Definio do Problema</h1>

<h2>Problema</h2>
<pre> Falta de dados com fcil acesso de maneira offline.</pre>

<h2>Causa</h2>
<pre> Dificuldade em acessar determinados dados.
     Dados em outras fontes incompletos e sem atualizaes.<pre>

<h2>Produto Final</h2>

     Aplicao no formato exe.<pre>
